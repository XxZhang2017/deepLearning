```python
# this script takes a folder full of source images and for each image makes a 
# bunch of slight variants. Those variants are put into the output dir.
import os
import shutil
from keras.preprocessing.image import ImageDataGenerator,img_to_array, load_img

def VarFile(file, in_dir, out_dir, num):
    if not os.path.exists(out_dir):
        os.makedirs(out_dir)

    datagen = ImageDataGenerator(
            width_shift_range=0.1,
            height_shift_range=0.1,
            horizontal_flip=True,
            fill_mode='nearest')

    full_in = os.path.join(in_dir, file)
    img = load_img(full_in)  
    x = img_to_array(img)  # creating a Numpy array with shape (3, 150, 150)
    x = x.reshape((1,) + x.shape)  # converting to a Numpy array with shape (1, 3, 150, 150)
    i = 0
    for batch in datagen.flow(x,save_to_dir=out_dir, save_prefix=file, save_format='ppm'):
        i += 1
        if i >= num:
            break

    shutil.copyfile(full_in, os.path.join(out_dir, file))  

        
```

```python
def MakeVariants(in_dir, out_dir, num):
    print('Make Variants', in_dir, out_dir, num)
    filenames = os.listdir(in_dir)
    for f in filenames:
        fullpath = os.path.join(in_dir, f)
        # recursively pick up subdirectories
        if os.path.isdir(fullpath):
            MakeVariants(fullpath, os.path.join(out_dir, f), num)
        else:
            if f.endswith(".ppm"):
                VarFile(f, in_dir, out_dir, num)
```

```python
inDir = "Marcel-Train"
outDir = "more-train"
MakeVariants(inDir, outDir, 3)

```
